```python
df = pd.read_csv('../data/dataproject2025.csv')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.head()
```

```python


class MultiColumnLabelEncoder(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.encoders = {}
        self.columns = None

    def fit(self, X, y=None):
        self.columns = X.columns
        for col in X.columns:
            le = LabelEncoder()
            X[col] = X[col].astype(str)
            le.fit(X[col])
            self.encoders[col] = le
        return self

    def transform(self, X):
        check_is_fitted(self, "encoders")
        X_out = X.copy()
        for col, le in self.encoders.items():
            X_out[col] = le.transform(X_out[col].astype(str))
        return X_out

    def get_feature_names_out(self, input_features=None):
        check_is_fitted(self, "encoders")
        return self.columns


# --- Targets ---
y = df['Predicted probabilities']
y_target = df['target']

# --- Features ---
X = df.drop(columns=['Predicted probabilities', 'Predictions', 'target'])
print(X.shape)
print(y.shape)
import numpy as np

def clean_emp_length(val):
    if pd.isnull(val):
        return np.nan
    val = str(val).lower()
    if '<' in val:
        return 0
    elif '10' in val:
        return 10
    elif 'n/a' in val:
        return np.nan
    else:
        return int(val.split()[0])

# Apply conversion
X['emp_length'] = X['emp_length'].apply(clean_emp_length)
# --- Column groups based on encoding table ---
std_scalar_cols = [
    'annual_inc','avg_cur_bal','bc_open_to_buy','bc_util','delinq_2yrs','dti',
    'emp_length','fico_range_high','funded_amnt','inq_last_6mths','int_rate',
    'mo_sin_old_rev_tl_op','mo_sin_rcnt_rev_tl_op','mo_sin_rcnt_tl',
    'mort_acc','mths_since_recent_bc','num_actv_bc_tl','num_bc_tl','num_il_tl',
    'num_rev_accts','open_acc','pub_rec','pub_rec_bankruptcies','revol_bal',
    'revol_util','tax_liens'
]

ordinal_cols = ['grade','sub_grade']
label_encoded_cols = ['emp_title','home_ownership','purpose']
no_encoding_cols = ['issue_d','zip_code','Pct_afro_american']

# --- Transformers ---
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

label_transformer = Pipeline(steps=[
    ('encoder', MultiColumnLabelEncoder())
])

ordinal_transformer = Pipeline(steps=[
    ('encoder', OrdinalEncoder())
])

# --- Preprocessor ---
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, std_scalar_cols),
        ('label', label_transformer, label_encoded_cols),
        ('ord', ordinal_transformer, ordinal_cols),
        ('passthrough', 'passthrough', no_encoding_cols)
    ]
)
```

```python
# Fit and transform
X_transformed = preprocessor.fit_transform(X)
feature_names = preprocessor.get_feature_names_out()
X_transformed_df = pd.DataFrame(X_transformed.toarray() if hasattr(X_transformed, "toarray") else X_transformed,
                                columns=feature_names, index=X.index)
X_transformed_df.head()
```

```python
df_preprocessed = pd.concat([X_transformed_df, y.reset_index(drop=True), y_target.reset_index(drop=True)], axis=1)
df_preprocessed.head()
```
